```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
dataset= [['Rendang', 'Nasi Putih', 'Sambal Ijo', 'Otak', 'Es Teh', 'Ayam Pop', 'Paru'],
         ['Paru', 'Kikil', 'Otak', 'Sayur Nangka', 'Nasi Putih', 'Es Teh', 'Ayam Pop', 'Telur Dadar'],
         ['Perkedel', 'Sayur Nangka', 'Kepala Kakap', 'Rendang', 'Bawal Bakar', 'Cumi', 'Ati', 'Es Teh'],
         ['Paru', 'Kikil', 'Otak', 'Sayur Nangka', 'Nasi Putih', 'Air', 'Ati', 'Telur Dadar'],
         ['Otak', 'Cumi', 'Ati', 'Sayur Nangka', 'Nasi Putih', 'Es Teh', 'Rendang', 'Daun Singkong'],
         ['Paru', 'Kikil', 'Otak', 'Sayur Nangka', 'Nasi Putih', 'Teh Tawar', 'Daun Singkong', 'Ayam Gulai'],
         ['Telur Dadar', 'Kikil', 'Perkedel', 'Sayur Nangka', 'Nasi Putih', 'Es Teh', 'Kepala Kakap', 'Babat'],
         ['Paru', 'Rendang', 'Daun Singkong', 'Sayur Nangka', 'Nasi Putih', 'Air', 'Sambal Ijo', 'Perkedel'],
         ['Air', 'Daun Singkong', 'Sayur Nangka', 'Nasi Putih', 'Es Teh', 'Cumi', 'Bawal Bakar'],
         ['Otak', 'Cumi', 'Kepala Kakap', 'Daun Singkong', 'Nasi Putih', 'Teh Tawar', 'Telur Dadar']]
```

```python
dataset
```

```python
#bikin data frame list barang yang dibeli dengan true/false
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
#menghitung support
frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.85)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.4)
rules
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
```

```python
rules[ (rules['antecedent_len'] >= 2) & (rules['confidence'] > 0.5) & (rules['lift'] > 1.4) ]
```
